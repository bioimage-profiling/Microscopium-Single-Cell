```python
fov_256_min_n = coordf.fov.unique().astype('uint8').min()
fov_64_min_n = centerdf.fov.unique().astype('uint8').min()
```

```python
B = np.arange(256)
A = B//4
A += fov_64_min_n
B += fov_256_min_n
fovdict = {}
for a, b in zip(A, B):
    fovdict[b] = a
```
